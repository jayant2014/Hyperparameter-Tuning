### Perform Data Cleaning Before Training

```python
# Check if there is any missing values
sns.heatmap(bike_df.isnull())
# No missing values in such
```

```python
# Drop instant, casual and registered as we are interested in total count only
bike_df = bike_df.drop(labels = ['instant', 'casual', 'registered'], axis = 1)
bike_df
```

```python
# Converting dteday to generic date format
bike_df['dteday'] = pd.to_datetime(bike_df['dteday'], format = '%m/%d/%Y')
bike_df
```

```python
# Update the index with datetime column
bike_df.index = pd.DatetimeIndex(bike_df['dteday'])
bike_df
```

```python
# Now since we have index ready which is asame as dteday, let's drop dteday
bike_df = bike_df.drop(labels = ['dteday'], axis = 1)
bike_df
```

### Let's do some exploratory data analysis

```python
# Let's plot bike rental usage per week
plt.figure(figsize = (12, 7))
bike_df['cnt'].asfreq('W').plot(linewidth = 5)
plt.title('Bike Rental Usage Per Week')
plt.xlabel('Week')
plt.ylabel('Bike Rental')
plt.grid()
```

```python
# PLot the same for monthly
plt.figure(figsize = (12, 7))
bike_df['cnt'].asfreq('M').plot(linewidth = 5)
plt.title('Bike Rental Usage Per Month')
plt.xlabel('Month')
plt.ylabel('Bike Rental')
plt.grid()
```

```python
# Plot the same for quarterly interval
bike_df['cnt'].asfreq('Q').plot(linewidth = 6)
plt.title('Bike Usage Per Quarter')
plt.xlabel('Quarter')
plt.ylabel('Bike Rental')
plt.grid()
```

```python
# Identify the numerical fields
X_numerical = bike_df[['temp', 'hum', 'windspeed', 'cnt']]
X_numerical
```

```python
# Pairplot the numerical features
sns.pairplot(X_numerical);
```

```python
# Let's see the correlation between numerical features
plt.figure(figsize = (8, 6))
sns.heatmap(X_numerical.corr(), annot = True);
```

```python
# Let's see the categorical fields
X_cat = bike_df[['season', 'yr', 'mnth', 'holiday', 'weekday', 'workingday', 'weathersit']]
X_cat
```

```python
# Apply one-hot encoding to categorical features
from sklearn.preprocessing import OneHotEncoder
onehotencoder = OneHotEncoder()
X_cat = onehotencoder.fit_transform(X_cat).toarray()
X_cat
```

```python
X_cat.shape
```

```python
# Convert categorical features into pandas dataframe
X_cat = pd.DataFrame(X_cat)
```

```python
X_numerical
```

```python
X_numerical = X_numerical.reset_index()
X_numerical
```

```python
# Apply numerical and categorical features
X_all = pd.concat([X_cat, X_numerical], axis = 1)
X_all
```

```python
# Drop dteday column
X_all = X_all.drop('dteday', axis = 1)
X_all
```

```python
# Prepare target feature data
X = X_all.iloc[:, :-1].values
y = X_all.iloc[:, -1:].values
```

```python
X.shape
```

```python
y.shape
```

### Training and Testing Dataset Split

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2)
```

```python
X_train.shape
```

```python
X_test.shape
```

### Training with XGBOOST Without Optimization

```python
import xgboost as xgb
model = xgb.XGBRegressor(objective ='reg:squarederror', learning_rate = 0.1, max_depth = 20, n_estimators = 500)
model.fit(X_train, y_train)
```

```python
# Predict the score of the trained model using the testing dataset
result = model.score(X_test, y_test)
print("Accuracy : {}".format(result))
```

```python
# Predictions on the test data
y_predict = model.predict(X_test)
```

```python
y_test
```

```python
y_predict
```

```python
# KPI measurements
from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error
from math import sqrt

k = X_test.shape[1]
n = len(X_test)
RMSE = float(format(np.sqrt(mean_squared_error(y_test, y_predict)),'.3f'))
MSE = mean_squared_error(y_test, y_predict)
MAE = mean_absolute_error(y_test, y_predict)
r2 = r2_score(y_test, y_predict)
adj_r2 = 1-(1-r2)*(n-1)/(n-k-1)

print('RMSE =',RMSE, '\nMSE =',MSE, '\nMAE =',MAE, '\nR2 =', r2, '\nAdjusted R2 =', adj_r2) 
```

### Model Optimization Using Grid Search

```python
from sklearn.model_selection import GridSearchCV
```

```python
parameters_grid = { 'max_depth': [3, 6, 10], 
                   'learning_rate': [0.01, 0.05, 0.1],
                   'n_estimators': [100, 500, 1000],
                   'colsample_bytree': [0.3, 0.7]}
# Total 54 combinations
```

```python
model = xgb.XGBRegressor(objective ='reg:squarederror')
```

```python
# We use the "neg_mean_squared_error" since GridSearchCV() ranks all the algorithms (estimators) 
# and specifies which one is the best 
xgb_gridsearch = GridSearchCV(estimator = model, 
                              param_grid = parameters_grid, 
                              scoring = 'neg_mean_squared_error',  
                              cv = 5, 
                              verbose = 5)
```

```python
xgb_gridsearch.fit(X_train, y_train)
```

```python
xgb_gridsearch.best_params_
```

```python
xgb_gridsearch.best_estimator_
```

```python
y_predict = xgb_gridsearch.predict(X_test)
```

```python
from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error
from math import sqrt
k = X_test.shape[1]
n = len(X_test)
RMSE = float(format(np.sqrt(mean_squared_error(y_test, y_predict)),'.3f'))
MSE = mean_squared_error(y_test, y_predict)
MAE = mean_absolute_error(y_test, y_predict)
r2 = r2_score(y_test, y_predict)
adj_r2 = 1-(1-r2)*(n-1)/(n-k-1)

print('RMSE =',RMSE, '\nMSE =',MSE, '\nMAE =',MAE, '\nR2 =', r2, '\nAdjusted R2 =', adj_r2) 
```

### Model Optimization Using Random Search

```python
from sklearn.model_selection import RandomizedSearchCV
```

```python
# Define the grid of hyperparameters to search. We can choose different booster
# There are two type of boosters - gbtree and gblinear
# gbtree uses tree based models while gblinear uses linear functions

grid = {
    'n_estimators': [100, 500, 700],
    'max_depth': [2, 3, 5],
    'learning_rate': [0.1, 0.5, 1],
    'min_child_weight': [1, 2, 3]  
            }

grid = {
    'n_estimators': [100, 500, 700],
    'max_depth': [2, 3, 5],
    'learning_rate': [0.1, 0.5, 1],
    'min_child_weight': [1, 2, 3],
    'booster': ['gbtree','gblinear']}

import xgboost as xgb
model = xgb.XGBRegressor(objective ='reg:squarederror')

random_cv = RandomizedSearchCV(estimator = model,
                               param_distributions = grid,
                               cv = 5, 
                               n_iter = 50,
                               scoring = 'neg_mean_absolute_error',
                               verbose = 5, 
                               return_train_score = True)
random_cv.fit(X_train, y_train)

random_cv.best_estimator_
```

```python
y_predict = random_cv.predict(X_test)
```

```python
from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error
from math import sqrt
k = X_test.shape[1]
n = len(X_test)
RMSE = float(format(np.sqrt(mean_squared_error(y_test, y_predict)),'.3f'))
MSE = mean_squared_error(y_test, y_predict)
MAE = mean_absolute_error(y_test, y_predict)
r2 = r2_score(y_test, y_predict)
adj_r2 = 1-(1-r2)*(n-1)/(n-k-1)

print('RMSE =',RMSE, '\nMSE =',MSE, '\nMAE =',MAE, '\nR2 =', r2, '\nAdjusted R2 =', adj_r2) 
```

```python
# Scikit-Optimize (Skopt) is used to perform bayesian optimization 
# Specify the search space as a distribution instead of discrete values

from skopt import BayesSearchCV
```

```python
model = xgb.XGBRegressor(objective ='reg:squarederror')
```

```python
search_space = {
    "learning_rate": (0.01, 1.0, "log-uniform"),
    "max_depth": (1, 50),
    "n_estimators": (5, 500),
}
```

```python
xgb_bayes_search = BayesSearchCV(model, 
                               search_space, 
                               n_iter = 50, 
                               scoring = 'neg_mean_absolute_error', 
                               cv = 5)
```

```python
result = xgb_bayes_search.fit(X_train, y_train)
```

```python
y_predict = xgb_bayes_search.predict(X_test)
```

```python
from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error
from math import sqrt
k = X_test.shape[1]
n = len(X_test)
RMSE = float(format(np.sqrt(mean_squared_error(y_test, y_predict)),'.3f'))
MSE = mean_squared_error(y_test, y_predict)
MAE = mean_absolute_error(y_test, y_predict)
r2 = r2_score(y_test, y_predict)
adj_r2 = 1-(1-r2)*(n-1)/(n-k-1)

print('RMSE =',RMSE, '\nMSE =',MSE, '\nMAE =',MAE, '\nR2 =', r2, '\nAdjusted R2 =', adj_r2) 
```
